# Demo - TDvsCPu
This demonstration shows the process of training, validating for typically developing (TD) and unilateral cerebral palsy (CPu) dataset using ResNet. At the end it also shows how it can be used to predict input cycles given to diagnose.

## Load the data and plot

```python


from utils.utils import *
import numpy as np

# Load dataset
dataset = "archives/AQM/TDvsPC_new_no_shuffle/"
d_file_list = ['CP.npz','TD.npz']
x_list, nb_classes = load_dataset_v1(dataset, d_file_list, channel_first=True,flatten=False)

# Load subject cycle table and split the dataset to training, validation and test set
idx_file_list = ['DataCP - By Subject.csv','DataTD - By Subject.csv']
x_train, x_test, y_train, y_test, x_val, y_val, cycle_end_idx, X_d, y_d = generate_data_for_train(x_list, idx_file_list,order=[1,0])

# Plot cycles of one subject
n_subject = 0
name_dataset = "TDvsCP"
RL_list = ['Affected Side','Unaffected Side']
plot_cycle_with_idx(n_subject, name_dataset, RL_list, cycle_end_idx, X_d)
```

## Train 10 models

```python
# Train 10 models

D_name = "TDvsCPu_new_full_5test"

for i in ['ResNet']:
    for j in range(10):
        train_model(D_name,Net_name=i, X_d=X_d,y_d=y_d,x_train=x_train,x_val=x_val,y_train=y_train,y_val=y_val,cycle_end_idx=cycle_end_idx)
```

## Evaluation 10 models by test set

```python
D_name = "TDvsCPu_new_full_5test"
Net_name = 'ResNet'

out = eval_model_5_bundle(D_name,Net_name,X_d,y_d,x_test,y_test,cycle_end_idx)
```

## Print the evaluation results of test set by cycles

```python
# The outputs are the Accuracy, Confusion matrix, F1 score, Sensitivity, Specificity and AUC for all of 10 model and their mean and STD.

_ = print_result_bundle_i_cycles(out,y_d,x_test,cycle_end_idx,label_list=['TD', 'CP'],n_model=10)
```

## Print the evaluation results of test set by subjects

```python
_ = print_result_bundle_i_subjects(out,y_d,x_test, y_test,cycle_end_idx,label_list=['TD', 'CP'],n_model=10)
```

## Prediction of new cycles

To evaluation new cycles to do a diagnosis, the new cycles can be added from input_data as the input for function predict_new_cycles. And the model_idx indicates which of the trained is models needs to be used for prediction. Here we use model 2 which has the highest accuracy.

```python
D_name = "TDvsCPu_new_full_5test"
Net_name = 'ResNet'
out = predict_new_cycles(input_data=X_d[:5], D_name=D_name,Net_name=Net_name,model_idx=2)
```

## The reuslt of prediction
class 1 is CPu

```python
print('prediction:',[np.argmax(i) for i in out])
```
